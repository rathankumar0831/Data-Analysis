# b. Find the which batsmen have been "caught and bowled" and how many have been “hit wicket” 

```python
## batsmen have been caught and bowled
filt1 = (df1['wicket_type'] == 'caught ') | (df1['wicket_type'] == 'bowled')
df1.loc[filt1,'striker'].count()
```

```python
## how many batsmen have been hit wicket
filt2 = df1['wicket_type'] == 'hit wicket'
df1.loc[filt2, 'striker'].count()
```
